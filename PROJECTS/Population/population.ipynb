# AGGREGATE VALUES IN DATA FRAME

- I have cross-referred the results with the original table of Malaysian population from Department of Statistics -- All figures are correct wiht this formula!ðŸ¥³ðŸ¥³ðŸ¥³

- General population for one whole year for various age brackets

- Grand total population for each year from 1991 to 2023
    * Filter data frame for the range of years youa re interested in
    * Group by 'Year' and sum up the 'Total' column for each year

- Grand total of male and female population 

- Combine total population, male population and female population in a new data frame pd.merge()
    * It can only handle two dataframes at a time, but it can be chained to combine more df


- note: we can also use pd.concat() since they share the same index and year. 

```python
#general population for 1991
general_pop = df_90.groupby(['Year', 'Age', 'Bumi', 'Chinese', 'Indian', 'Others']).agg({
    'Total': 'sum'
}).reset_index()

print(general_pop.head(18))
```

```python
#grand total_population
grand_total_population = df_90.groupby('Year').agg({
    'Total': 'sum'
}).reset_index()

grand_total_population
```

```python
#Grand total of male population
df_90['Male_Total'] = df_90[["Male_Bumi", "Male_Chinese", "Male_Indian", "Male_Others",	"Male_non Citizens"]].sum(axis=1)

#Group by 'Year' and sum of the 'Male_Total' column
grand_total_males = df_90.groupby('Year').agg({
    'Male_Total': 'sum'
}).reset_index()

grand_total_males
```

```python
#Grand total of female population
df_90['Female_Total'] = df_90[["Female_Bumi", "Female_Chinese", "Female_Indian", "Female_Others", "Female_non Citizens"]].sum(axis=1)

#Group by 'Year' and sum of the 'Male_Total' column
grand_total_females = df_90.groupby('Year').agg({
    'Female_Total': 'sum'
}).reset_index()

grand_total_females
```

**COMBINE 3 NEW DATAFRAMES INTO A NEW ONE**

1) Grand total population
2) grand total male
3) grand total female

```python
combined_pop = pd.merge(grand_total_population, grand_total_males, on='Year')

combined_pop = pd.merge(combined_pop, grand_total_females, on='Year')

#rename column
combined_pop.columns = ['Year', 'Total Population', 'Total Male', 'Total Female']

combined_pop
```
