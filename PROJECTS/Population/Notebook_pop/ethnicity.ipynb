# Aggregate values for ethnicity across male population



```python
#Total population across years
Total_pop = big_data.groupby('Year').agg({
    'Total': 'sum'
}).reset_index()

#total of each ethncity across male population
Male_bumi_tot = big_data.groupby('Year').agg({
    'Male_Bumi': 'sum'
}).reset_index()

Male_chi_tot = big_data.groupby('Year').agg({
    'Male_Chinese': 'sum'
}).reset_index()

Male_ind_tot = big_data.groupby('Year').agg({
    'Male_Indian': 'sum'
}).reset_index()

Male_oth_tot = big_data.groupby('Year').agg({
    'Male_Others': 'sum'
}).reset_index()

Male_NC_tot = big_data.groupby('Year').agg({
    'Male_non Citizens': 'sum'
}).reset_index()

Male_NC_tot.head()
```

**SIMPLIFIED AGGREGATION - MALE + RACE**

```python
total_eth_male = big_data.groupby('Year').agg({
    'Total': 'sum',
    'Male_Bumi': 'sum',
    'Male_Chinese': 'sum',
    'Male_Indian': 'sum',
    'Male_Others': 'sum',
    'Male_non Citizens': 'sum'
}).reset_index()

total_eth_male.head()
```

**SIMPLIFIED AGGREGATION - FEMALE + RACE**

```python
total_eth_fem = big_data.groupby('Year').agg({
    'Total': 'sum',
    'Female_Bumi': 'sum',
    'Female_Chinese': 'sum',
    'Female_Indian': 'sum',
    'Female_Others': 'sum',
    'Female_non Citizens': 'sum'
}).reset_index()

total_eth_fem.head()
```

**combined aggregation across years + ethnicity + male & female

```python
total_eth_male = total_eth_male.drop(columns='Total')
total_eth_fem = total_eth_fem.drop(columns='Total')

comb_eth_FM = total_eth_male.merge(total_eth_fem, on='Year')

comb_eth_FM.head()
```
